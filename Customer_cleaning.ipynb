```python
# Import Library
import numpy as np
import pandas as pd
```

```python
# Import the dataset
data = pd.read_csv("D:\DA\DA PROJECT\DA DATA\Project 4\dbo.customer.csv", encoding='utf-8')
data1 = pd.read_csv("D:\DA\DA PROJECT\DA DATA\Project 4\dim.date.csv", encoding='utf-8')
data2 = pd.read_csv("D:\DA\DA PROJECT\DA DATA\Project 4\dim.geography.csv", encoding='utf-8')
data3 = pd.read_csv("D:\DA\DA PROJECT\DA DATA\Project 4\dim.sales_territory.csv", encoding='utf-8')
data4 = pd.read_csv("D:\DA\DA PROJECT\DA DATA\Project 4\\fact_internet_sales.csv", encoding='utf-8')
data5 = pd.read_csv("D:\DA\DA PROJECT\DA DATA\Project 4\dim.product.csv", encoding='utf-8')

# Import dataset into list
data_list = [data, data1, data2, data3, data4, data5]

# Display the first 5 rows of the DataFrame in the list
for i, df in enumerate(data_list):
    print(f"DataFrame {i+1}:\n", df.head(), "\n")
```

```python
# Eliminate unnecessary columns each dataframe in the list
for i, df in enumerate(data_list):
    if i == 0:  # dbo.customer
        data_list[i] = df[['CustomerKey', 'GeographyKey', 'FirstName', 'LastName', 'BirthDate', 'MaritalStatus', 'Gender', 'EmailAddress', 'Phone']]

    elif i == 1:  # dim.date
        #  No columns to eliminate
        data_list[i] = data_list[i] 
    elif i == 2:  # dim.geography
        data_list[i] = df[['GeographyKey', 'City', 'EnglishCountryRegionName']]
    elif i == 3:  # dim.sales_territory
        data_list[i] = data_list[i]
    elif i == 4:  # fact_internet_sales
        data_list[i] = data_list[i]
    elif i == 5:  # dim.product
        data_list[i] = df[['ProductKey', 'EnglishProductName']]
```

```python
#Check for missing values in each DataFrame
for i, df in enumerate(data_list):
    missing_values = df.isnull().sum()
    print(f"Missing values in DataFrame {i+1}:\n{missing_values}\n")
```

```python
# Check for duplicate rows in each DataFrame
for i, df in enumerate(data_list):
    duplicate_rows = df.duplicated().sum()
    print(f"Number of duplicate rows in DataFrame {i+1}: {duplicate_rows}\n")
```

```python
# Delete extra spaces in string columns in each DataFrame
for i, df in enumerate(data_list):
    df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
    data_list[i] = df # Update the DataFrame in the list
# Display the cleaned DataFrames
for i, df in enumerate(data_list):
    print(f"Cleaned DataFrame {i+1}:\n", df.head(), "\n")
```
